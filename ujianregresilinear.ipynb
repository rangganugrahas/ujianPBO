```python
#manipulasi data
import pandas as pd

#PATH ATAU PEMANGGIL
path ='/content/drive/MyDrive/CSV/indonesia - Sheet1.csv' 

#DF DATA FRAME /UNTUK PENAMAAN BEBAS
df = pd.read_csv(path) 

#melihat data struktur
df.describe() 
```

```python
df
```

```python
#VISUALISASI DATA
import matplotlib.pyplot as plt 

df.plot(kind='scatter', x='Life expectancy', y='percentage expenditure')#untuk nama variable

plt.title('Data Indonesia')
plt.xlabel('Life Expetacy')
plt.ylabel('Percentage expenditure')
plt.xlim(66.3, 69.1)
plt.ylim(0.10, 8.46)
plt.grid(True)
plt.show()
```

```python
#VISUALISASI DATA
import matplotlib.pyplot as plt 

df.plot(kind='scatter', x='Population', y='GDP')#untuk nama variable

plt.title('Data Indonesia')
plt.xlabel('Life Expetacy')
plt.ylabel('Percentage expenditure')
plt.xlim(2, 248)
plt.ylim(0.1, 8.99)
plt.grid(True)
plt.show()
```
